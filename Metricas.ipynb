```python
import matplotlib.pyplot as plt
from facedetector import FaceDetector
import cv2

import os
import glob
import random

def get_first_n_images(directory, n=50):
    image_files = glob.glob(os.path.join(directory, '*.jpg'))
    return random.sample(image_files, n)


first_50_images = get_first_n_images('./imagenes/part1/')

faceDetector = FaceDetector()
detections = [
    cv2.cvtColor(faceDetector.processSingleImage(img), cv2.COLOR_RGB2BGR)
    for img in first_50_images
]

fig, axes = plt.subplots(5, 10, figsize=(20, 10))

for i, ax in enumerate(axes.flatten()):    
    image = detections[i]
    ax.imshow(image)
    ax.axis('off')
    
plt.show()
```

En base a las 50 imagenes de prueba se tiene:
- True positive (TP): 44
- False positive (FP): 3
- False negative (FP): 6

```python
def calculate_metrics(TP, FP, FN):
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    jaccard_index = TP / (TP + FP + FN)
    return precision, recall, jaccard_index

# Datos
TP = 44
FP = 3
FN = 6

# Calcular métricas
precision, recall, jaccard_index = calculate_metrics(TP, FP, FN)

# Mostrar la tabla de métricas
print("Métricas:")
print("----------")
print("Precision:", precision)
print("Recall:", recall)
print("Jaccard Index:", jaccard_index)
```
